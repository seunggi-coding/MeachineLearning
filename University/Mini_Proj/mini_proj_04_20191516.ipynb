import numpy as np
import pandas as pd    
from itertools import combinations
from pycaret.regression import *
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, VotingRegressor
from sklearn.model_selection import cross_val_score

input_file = "winequality-red.csv"
train = pd.read_csv(input_file)

wine_data = train.drop("quality", axis=1)
wine_labels = train["quality"].copy()

# k-폴드 CV의 평균을 계산하기 위한 함수
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())


# 데이터프레임에서 특성 추출
features = train[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 
                   'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']]

# 모든 특성의 조합을 만들기
combinations_list = list(combinations(features.columns, 2))

# 조합된 특성들의 평균을 나타내는 새로운 특성들 생성
for combo in combinations_list:
    feature_name = f"({combo[0]}+{combo[1]})/2"
    train[feature_name] = (features[combo[0]] + features[combo[1]]) / 2
    
    feature_name2 = f"({combo[0]}-{combo[1]})/2"
    train[feature_name2] = (features[combo[0]] - features[combo[1]]) / 2
    
    feature_name3 = f"({combo[0]}*{combo[1]})/2"
    train[feature_name3] = (features[combo[0]] * features[combo[1]]) / 2
    
    feature_name4 = f"({combo[0]}/{combo[1]})/2"
    train[feature_name4] = (features[combo[0]] / features[combo[1]]) / 2
    
    feature_name5 = f"({combo[0]}+{combo[1]})"
    train[feature_name5] = (features[combo[0]] + features[combo[1]])
    
    feature_name6 = f"({combo[0]}-{combo[1]})"
    train[feature_name5] = (features[combo[0]] - features[combo[1]])
    
    feature_name7 = f"({combo[0]}*{combo[1]})"
    train[feature_name7] = (features[combo[0]] * features[combo[1]])
    
    feature_name8 = f"({combo[0]}/{combo[1]})"
    train[feature_name5] = (features[combo[0]] / features[combo[1]])

corr_matrix = train.corr()
quality = corr_matrix['quality']
quality.describe()

filtered_train_data = train.drop(quality[quality <  0.17].index, axis=1)
filtered_train_data = filtered_train_data.drop("quality", axis=1)
filtered_train_data.count()

train_target = train["quality"]

# RF
rf_regressor = RandomForestRegressor(random_state=42)

# Ada
ada_regressor = AdaBoostRegressor(random_state=42)

# Voting Regressor 생성
voting_regressor = VotingRegressor(
    estimators=[('rf', rf_regressor  ), ('ada', ada_regressor)]
)

# Voting Regressor 학습
voting_regressor.fit(filtered_train_data, train_target)

scores = cross_val_score(voting_regressor, filtered_train_data, train_target, scoring="neg_mean_squared_error", cv=10, n_jobs=-1)
rmse_scores = np.sqrt(-scores)
display_scores(rmse_scores)

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.model_selection import cross_val_score

# 데이터 불러오기
input_file = "winequality-red.csv"
data = pd.read_csv(input_file)
tmp = data
data = data.drop("residual sugar", axis=1)
data = data.drop("density", axis=1)

# 추가적인 계산
data['total_acidity'] = data['fixed acidity'] + data['volatile acidity']
data['average_acidity'] = (data['fixed acidity'] + data['volatile acidity']) / 2
data['total_minerals'] = data['chlorides'] + data['sulphates']
data['average_minerals'] = (data['chlorides'] + data['sulphates']) / 2
data['non_free_sulfur_dioxide'] = np.abs(data['total sulfur dioxide'] - data['free sulfur dioxide'])

# percentage_free_sulfur 계산
data['percentage_free_sulfur'] = data['free sulfur dioxide'] / data['total sulfur dioxide']
data['percentage_free_sulfur'].replace([np.inf, -np.inf], np.nan, inplace=True)
data['percentage_free_sulfur'].fillna(0, inplace=True)

# 추가적인 계산
data['percentage_salt_sulfur'] = data['sulphates'] / data['free sulfur dioxide']
data['percentage_salt_sulfur'].replace([np.inf, -np.inf], np.nan, inplace=True)
data['percentage_salt_sulfur'].fillna(0, inplace=True)

# 더 많은 계산
data["balance"] = data["alcohol"] - data["sulphates"] - data["volatile acidity"]
data['total_minerals'] = data['chlorides'] + data['sulphates']
data['sulph_sub_chlor']= data['sulphates']- data['chlorides']

data['(2citric acid+alcohol)'] = 2*data['citric acid'] + data['alcohol']

data['1'] = data['balance'] + 2*data['sulph_sub_chlor']
data['2'] = data['volatile acidity']**data['sulph_sub_chlor']
data['3'] = (data['alcohol']/data['1'])

# 목표 변수 및 불필요한 열 제거
train_data = data.drop('quality', axis=1)
train_data = train_data.dropna(axis=1)
train_target = data['quality']

# StandardScaler를 사용하여 데이터 표준화
scaler = StandardScaler()
scaler.fit(train_data)
train_data = scaler.transform(train_data)

# k-폴드 CV의 평균을 계산하기 위한 함수
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())

# 데이터 섞기
seed = 553419
np.random.seed(seed)
sh_in = np.random.permutation(len(train_data))
sh_t_data = train_data[sh_in]
sh_t_target = train_target[sh_in]

# ExtraTreesRegressor 모델 초기화
extra_trees_clf = ExtraTreesRegressor(max_depth=30, max_features=3, min_samples_leaf=1, min_samples_split=2,
                                      n_estimators=1500, random_state=42)

# 교차 검증 수행
scores = cross_val_score(extra_trees_clf, sh_t_data, sh_t_target, scoring="neg_mean_squared_error", cv=10, n_jobs=-1)
rmse_scores = np.sqrt(-scores)
print("===================================================================================================")
display_scores(rmse_scores)
print("===================================================================================================")
```
